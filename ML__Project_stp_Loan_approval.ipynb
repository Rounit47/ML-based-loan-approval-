From this output we are getting a mixture of integers and strings records, and one of the records with 'A' as well.

```python
# creating a function for str recodrs for formatting

def remove_str(x):
    if isinstance(x, str):
        return x.replace('A', '')
    return x
```

```python
# now calling the formatting function and make str records as int recodrs

df['ApprovalFY'] = df['ApprovalFY'].apply(remove_str).astype('int64')
```

Changing the same for few more features/columns as requirement

```python
# changing the dtypes of NewExist To intger, Zip and UrbanRural to str(categorical) and all currency related features to float values

df = df.astype({'Zip':'str', 'NewExist': 'int64', 'UrbanRural': 'str', 'DisbursementGross': 'float', 'BalanceGross': 'float',
                          'ChgOffPrinGr': 'float', 'GrAppv': 'float', 'SBA_Appv': 'float'})
```

```python
# checking data types for the changes
df.dtypes
```

Now we will address the NAICS feature next, which tells about the industry each business belongs to. we will use only the first two digits of each business NAICS code. we are also provided with the list of industries for same having corresponding numbers each. we will use it as reference.

```python
# creating a new column
# selecting only first two no. as NAICS code

df['Industry'] = df['NAICS'].astype('str').apply(lambda x: x[:2])
```

```python
# mapping the approprate industry to each records based on there NAISC code(only two digits)


df['Industry'] = df['Industry'].map({
    '11': 'Ag/For/Fish/Hunt',
    '21': 'Min/Quar/Oil_Gas_ext',
    '22': 'Utilities',
    '23': 'Construction',
    '31': 'Manufacturing',
    '32': 'Manufacturing',
    '33': 'Manufacturing',
    '42': 'Wholesale_trade',
    '44': 'Retail_trade',
    '45': 'Retail_trade',
    '48': 'Trans/Ware',
    '49': 'Trans/Ware',
    '51': 'Information',
    '52': 'Finance/Insurance',
    '53': 'RE/Rental/Lease',
    '54': 'Prof/Science/Tech',
    '55': 'Mgmt_comp',
    '56': 'Admin_sup/Waste_Mgmt_Rem',
    '61': 'Educational',
    '62': 'Healthcare/Social_assist',
    '71': 'Arts/Entertain/Rec',
    '72': 'Accom/Food_serv',
    '81': 'Other_no_pub',
    '92': 'Public_Admin'
})


df.dropna(subset=['Industry'], inplace=True)
```

```python
# checking first five records to see new feature industry is working our not

df.head()
```

Creating new feature 'IsFranchise' for storing binary values

```python
df.loc[(df['FranchiseCode'] <= 1), 'IsFranchise'] = 0
df.loc[(df['FranchiseCode'] > 1), 'IsFranchise'] = 1
```

Next we go for NewExist column

```python
# getting unique values of NewExist feature using unique function
df['NewExist'].unique()
```

```python
# Keep records where NewExist == 1 or 2
df = df[(df['NewExist'] == 1) | (df['NewExist'] == 2)]

# Create NewBusiness field where 0 = Existing business and 1 = New business; based on NewExist field

df.loc[(df['NewExist'] == 1), 'NewBusiness'] = 0

df.loc[(df['NewExist'] == 2), 'NewBusiness'] = 1
```

```python
# again using head function

df.head()
```

**Now we check for RevLineCr and LowDoc**

```python
# Double check RevLineCr and LowDoc unique values
df['RevLineCr'].unique()
```

```python
df['LowDoc'].unique()
```

```python
# Removing records where RevLineCr != 'Y' or 'N' and LowDoc != 'Y' or 'N'
df = df[(df['RevLineCr'] == 'Y') | (df['RevLineCr'] == 'N')]
df = df[(df['LowDoc'] == 'Y') | (df['LowDoc'] == 'N')]
```

```python
# RevLineCr and LowDoc: 0 = No, 1 = Yes

df['RevLineCr'] = np.where(df['RevLineCr'] == 'N', 0, 1)
df['LowDoc'] = np.where(df['LowDoc'] == 'N', 0, 1)

# Checking that it worked

print(df['RevLineCr'].unique())
print(df['LowDoc'].unique())
```


**Now Most Important we check the same for MIS_Status column**

```python
# Makeing  Default target column based on MIS_Status where P I F = 0 and CHGOFF = 1 so we can see what features are prevalant in a defaulted loan
df['Default'] = np.where(df['MIS_Status'] == 'P I F', 0, 1)
df['Default'].value_counts()
```

**Now we check for ApprovalDate and DisbursementDate columns**

```python
# Converting ApprovalDate and DisbursementDate columns to datetime values

df['ApprovalDate'] = pd.to_datetime(df['ApprovalDate'])
df['DisbursementDate'] = pd.to_datetime(df['DisbursementDate'])
```

```python
# checking Data types

df.dtypes
```

Now we will check that how many days it took from the approval of the loan to the actual disbursement of the funds.

```python
# Creating DaysToDisbursement column which calculates the total number of days  between DisbursementDate and ApprovalDate

df['DaysToDisbursement'] = df['DisbursementDate'] - df['ApprovalDate']
```

```python
df['DaysToDisbursement'] = df['DaysToDisbursement'].astype('str').apply(lambda x: x[:x.index('d') - 1]).astype('int64')
```

```python
# Create DisbursementFY field for time selection criteria later
df['DisbursementFY'] = df['DisbursementDate'].map(lambda x: x.year)
```

**We look whether or not the bank servicing the loan was in the same state that the business was located.**

```python
#  creating SameState column which tells where the business State is the same as the BankState
df['SameState'] = np.where(df['State'] == df['BankState'], 1, 0)
```

```python
 #Create SBA_AppvPct field since the guaranteed amount is based on a percentage of the gross loan amount rather than dollar amount in most situations
df['SBA_AppvPct'] = df['SBA_Appv'] / df['GrAppv']
```

I wish to look that whether the loan amount disbursed was equal to the full amount approved, so I added the feature for same.

```python
# Creating AppvDisbursed column signifying if the loan amount disbursed was equal to the full amount approved

df['AppvDisbursed'] = np.where(df['DisbursementGross'] == df['GrAppv'], 1, 0)
```

```python
# We check Data type again
df.dtypes
```

```python
# changing the data types of IsFranchise and NewBusiness column

df = df.astype({'IsFranchise': 'int64', 'NewBusiness': 'int64'})
df.dtypes
```

**Droping the columns which will have no impact on the model training**

```python
df.drop(columns=['LoanNr_ChkDgt', 'Name', 'City', 'Zip', 'Bank', 'NAICS', 'ApprovalDate', 'NewExist', 'FranchiseCode',
                      'ChgOffDate', 'DisbursementDate', 'BalanceGross', 'ChgOffPrinGr', 'SBA_Appv', 'MIS_Status'], inplace=True)
```

 I wanted to look at are whether a loan was backed by Real Estate, and whether a loan was active during the Great Recession (2007-2009).

 To determine whether a loan was backed by Real Estate, I made a flag that signifies if the loan term is >= 20 years

 For loans active during the Great Recession, I created a flag for loans where the Great Recession (2007-2009) between DisbursementFY and DisbursementFY plus the loan term (in years).

```python
# Field for loans backed by Real Estate (loans with a term of at least 20 years)
df['RealEstate'] = np.where(df['Term'] >= 240, 1, 0)

# Field for loans active during the Great Recession (2007-2009)
df['GreatRecession'] = np.where(((2007 <= df['DisbursementFY']) & (df['DisbursementFY'] <= 2009)) |
                                     ((df['DisbursementFY'] < 2007) & (df['DisbursementFY'] + (df['Term']/12) >= 2007)), 1, 0)
```

```python
# Select only records with a disbursement year through 2010
df = df[df['DisbursementFY'] <= 2010]
```

After data preprocessing and data visulaiztion the next important step is model training in which we train our model using different algorithm and chose the best algo for prediction.

**Importing Libraries for model Training**

```python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import time
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.model_selection import GridSearchCV
```

**Here i am using one hot encoding method**

```python
# One-hot encode categorical data
df= pd.get_dummies(df)

df.head()
```

```python
# Establish target and feature fields
y = df['Default']
x = df.drop('Default', axis=1)
```

**Scaleing the data and the Splitting the Training Data in train set and test set**

```python
# Scale the feature values prior to modeling
scale = StandardScaler()
x_scaled = scale.fit_transform(x)

```

```python
# Splitting data using Train_test_split
# Our 75% data will be train set and remaining 25% will be Test set and seed value is 7
seed=7
x_train, x_val, y_train, y_val = train_test_split (x_scaled, y, test_size = 0.98, random_state=seed)
```

Testing Different Models I will be testing the following models with my training data (got the list from here):

1) Logistic Regression.

2) Linear Discriminant Analysis(LDA).

3) k-Nearest Neighbors (kNN).

4) Naive Bayes (=NB = GaussianNB).

5)Classification and Regression Trees
(CART).

6)Support Vector Machines (svm)
(SVC-Support Vector Classifier)

7)RandomForestClassifier

For each model, we set the model, fit it with 75% of our training data, predict for 25% of the training data and check the accuracy. we will check start and end time as well for all model. we will traing model one by one

```python
# Model-1  Logistic Regression

logreg = LogisticRegression()
logreg.fit(x_train, y_train)
startTime = time.time()
y_pred = logreg.predict(x_val)
endTime = time.time()
acc_logreg = round(accuracy_score(y_pred, y_val) * 100, 2)
print("MODEL-1: Accuracy of LogisticRegression :      ", acc_logreg,)
print("(run time: %f)" % (endTime-startTime))
```

```python
# Model-2 Linear Discriminant Analysis(LDA).
lda=LinearDiscriminantAnalysis()
lda.fit(x_train, y_train)
startTime = time.time()
y_pred = lda.predict(x_val)
endTime = time.time()
acc_lda= round(accuracy_score(y_pred, y_val) * 100, 2)
print("MODEL-2: Accuracy of lda : ", acc_lda)
print("(run time: %f)" % (endTime-startTime))
```

```python
# Model-3 GaussianNB  .
gaussian = GaussianNB()
gaussian.fit(x_train, y_train)
startTime = time.time()
y_pred = gaussian.predict(x_val)
endTime = time.time()
acc_gaussian = round(accuracy_score(y_pred, y_val) * 100, 2)
print("MODEL-3: Accuracy of GaussianNB : ", acc_gaussian)
print("(run time: %f)" % (endTime-startTime))
```

```python
# MODEL-4) Decision Tree Classifier

decisiontree = DecisionTreeClassifier()
decisiontree.fit(x_train, y_train)
startTime = time.time()
y_pred = decisiontree.predict(x_val)
endTime = time.time()
acc_decisiontree = round(accuracy_score(y_pred, y_val) * 100, 2)
print("MODEL-4: Accuracy of DecisionTreeClassifier : ", acc_decisiontree)
print("(run time: %f)" % (endTime-startTime))
```

```python
# MODEL-5) knn
knn = KNeighborsClassifier(n_neighbors=3, n_jobs=-1)
knn.fit(x_train, y_train)
startTime = time.time()
y_pred = knn.predict(x_val)
endTime = time.time()
acc_knn = round(accuracy_score(y_pred, y_val) * 100, 2)
print("MODEL-5: Accuracy of KNN : ", acc_knn)
print("(run time: %f)" % (endTime-startTime))
```

```python
# MODEL-6) Random Forest


randomforest = RandomForestClassifier()
randomforest.fit(x_train, y_train)

startTime = time.time()
y_pred = randomforest.predict(x_val)
endTime = time.time()

acc_randomforest = round(accuracy_score(y_pred, y_val) * 100, 2)
print("MODEL-6: Accuracy of RandomForestClassifier:", acc_randomforest)
print("(run time: %f)" % (endTime - startTime))
```
